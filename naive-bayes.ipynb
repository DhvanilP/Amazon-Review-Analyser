```python
from __future__ import division #To avoid integer division
from operator import itemgetter
import pandas as pd
```

```python
data = pd.read_csv("data_set.csv", low_memory=False)
```

```python
data = data[['reviews.rating', 'reviews.text']]
```

```python
pos_rev = []
neg_rev = []
neut_rev = []
```

```python
# for i in range(len(data)):
#     if data[]
data['reviews.rating'] = pd.to_numeric(data['reviews.rating'], errors='coerce')
data = data.dropna(subset=['reviews.rating'])
data['reviews.rating'] = data['reviews.rating'].astype(int)

# a = 0
for col, row in data.iterrows():
    if int(row["reviews.rating"]) > 3:
        pos_rev.append(row["reviews.text"])
    elif int(row["reviews.rating"]) < 3:
        neg_rev.append(row["reviews.text"])
    else:
        neut_rev.append(row["reviews.text"])
# a
# neut_rev
```

```python
words = {}
pos_len = 0
neg_len = 0
neut_len = 0

# print (len(pos_rev))

for rev in pos_rev:
    try:
        spl_rev = rev.split()
        pos_len += 1
        for item in spl_rev:
            if item in words:
                words[item][0] += 1
            else:
                lis = [0] * 3
                words[item] = lis
                words[item][0] += 1
    except:
        print("h")

# pos_len

for rev in neg_rev:
    try:
        spl_rev = rev.split()
        pos_len += 1
        for item in spl_rev:
            if item in words:
                words[item][1] += 1
            else:
                lis = [0] * 3
                words[item] = lis
                words[item][1] += 1
    except:
        print("h")
        
for rev in pos_rev:
    try:
        spl_rev = rev.split()
        pos_len += 1
        for item in spl_rev:
            if item in words:
                words[item][0] += 1
            else:
                lis = [0] * 3
                words[item] = lis
                words[item][0] += 1
    except:
        print("h")
```
